```python
df_cleanedgpw3=pd.read_csv('cleaned_combined_dataset_gpw3.csv')
```

```python
df_cleanedgpw3.head()
```

```python
#graphe air pollution
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(10, 6))

# Plot 'air_pollution_average_[ug/m3]' vs 'populated_area_[km2]'
sns.lineplot(x='populated_area_[km2]', y='air_pollution_average_[ug/m3]', data=df_cleanedgpw3, label='Average Air Pollution [ug/m3]', marker='o')

# Plot 'air_pollution_population_weighted_average_[ug/m3]' vs 'populated_area_[km2]'
sns.lineplot(x='populated_area_[km2]', y='air_pollution_population_weighted_average_[ug/m3]', data=df_cleanedgpw3, label='Population Weighted Air Pollution [ug/m3]', marker='o')

# Add titles and labels
plt.title('Air Pollution Metrics by Populated Area')
plt.xlabel('Populated Area [km]')
plt.ylabel('Air Pollution [g/m]')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()

#The average air pollution per area seems affected by the population density until approximately 5000inhabitants/km but not in more dense area.
```
